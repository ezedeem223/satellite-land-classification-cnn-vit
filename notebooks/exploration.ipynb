# Exploration Notebook

This lightweight notebook demonstrates how the refactored package can be used without relying on the original course environment. It does not retrain models by default; instead it inspects the local configuration and dataset state.

```python
from pathlib import Path

from satellite_land_classification.config import load_config
from satellite_land_classification.data import build_image_index, validate_dataset_dir

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

data_config = load_config(repo_root / "configs" / "data.yaml")
dataset_dir = repo_root / data_config["dataset"]["extracted_dir"]
dataset_dir
```

```python
if dataset_dir.exists():
    validate_dataset_dir(dataset_dir)
    image_index = build_image_index(dataset_dir)
    print(f"Dataset is available at: {dataset_dir}")
    print(f"Number of indexed images: {len(image_index)}")
    print("First 3 examples:")
    for path, label in image_index[:3]:
        print(label, path)
else:
    print("Dataset not found locally. Run scripts/run_prepare_data.py first.")
```
